##1. Discount Level Prediction:

```python
print(df[['discount_percentage']].isnull().sum())

df['discount_percentage'] = df['discount_percentage'].str.rstrip('%').astype('float') / 100.0
print(df['discount_percentage'])

df['discount_percentage'].fillna(df['discount_percentage'].mean(), inplace=True)
```

```python
df['discount_percentage'].dtype
```

```python
high_discount_threshold = 0.2  # Set your own threshold
df['high_discount'] = (df['discount_percentage'] > high_discount_threshold).astype(int)
print(df['high_discount'])
```

```python
columns_to_drop = ['img_link', 'product_link', 'review_id', 'user_name']
df_discount = df.drop(columns=columns_to_drop)
```

```python
df_discount.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import OneHotEncoder

X_discount = df_discount.drop(['high_discount', 'product_id', 'product_name'], axis=1)
y_discount = df_discount['high_discount']

categorical_cols = X_discount.select_dtypes(include=['object']).columns
X_discount_encoded = pd.get_dummies(X_discount, columns=categorical_cols, drop_first=True)

X_train_discount, X_test_discount, y_train_discount, y_test_discount = train_test_split(
    X_discount_encoded, y_discount, test_size=0.2, random_state=42
)

model_discount = LogisticRegression()
model_discount.fit(X_train_discount, y_train_discount)

predictions_discount = model_discount.predict(X_test_discount)
accuracy_discount = accuracy_score(y_test_discount, predictions_discount)
print(f"Accuracy for Discount Level Prediction: {accuracy_discount}")
print("Classification Report:")
print(classification_report(y_test_discount, predictions_discount))
```

```python
coefficients = model_discount.coef_[0]
feature_names = X_train_discount.columns

coef_df = pd.DataFrame({'Feature': feature_names, 'Coefficient': coefficients})
coef_df = coef_df.sort_values(by='Coefficient', ascending=False)
print(coef_df)
```

### 24. Sentiment Analysis:

```python
df.head()
```

```python
print(df[['review_content']].isnull().sum())

df['review_content'].fillna('', inplace=True)
```

```python
df['review_content'] = df['review_content'].str.lower()
df['review_content'] = df['review_content'].apply(lambda x: ' '.join([word for word in x.split() if word.isalnum()]))
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

df['rating'].fillna(df['rating'].mean(), inplace=True)

positive_sentiment_threshold = 4  # Set your own threshold
df['positive_sentiment'] = (df['rating'] >= positive_sentiment_threshold).astype(int)
```

```python
X_sentiment = df.drop(['positive_sentiment', 'user_id', 'user_name', 'review_id', 'review_title', 'img_link', 'product_link'], axis=1)
y_sentiment = (df['rating'] >= 4).astype(int)

categorical_cols_sentiment = X_sentiment.select_dtypes(include=['object']).columns
X_sentiment_encoded = pd.get_dummies(X_sentiment, columns=categorical_cols_sentiment, drop_first=True)

X_train_sentiment, X_test_sentiment, y_train_sentiment, y_test_sentiment = train_test_split(
    X_sentiment_encoded, y_sentiment, test_size=0.2, random_state=42
)

model_sentiment = LogisticRegression()
model_sentiment.fit(X_train_sentiment, y_train_sentiment)

predictions_sentiment = model_sentiment.predict(X_test_sentiment)
accuracy_sentiment = accuracy_score(y_test_sentiment, predictions_sentiment)
print(f"Accuracy for Sentiment Analysis: {accuracy_sentiment}")
print("Classification Report:")
print(classification_report(y_test_sentiment, predictions_sentiment))
```
